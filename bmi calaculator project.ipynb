```python
name = input("Enter your name: ")

weight = int(input("Enter your weight in pounds: "))

height = int(input("Enter your height in inches: "))

BMI = (weight * 703) / (height * height)
print(f"{name} ur BMI is {BMI}")

if BMI>0:
    if (BMI<18.5):
       print(f"{name} you are umderweight")
    elif (BMI<=24.9):
        print(f"{name} you have a normal weight")
    elif (BMI<=29.9):
        print(f"{name} you have are overweight")
    elif (BMI<=34.9):
        print(f"{name} you have are obese")
    elif (BMI <=24.9):
        print(f"{name} your obessity level is high") 
    else:
        print(f"{name} your obessity level is extremely high")    
else:
   print("input valid value")
```
